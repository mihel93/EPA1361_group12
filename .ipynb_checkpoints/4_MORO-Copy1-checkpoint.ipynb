### 4. Multi-Objective Robust Optimization (MORO)

In notebook "3_comparison", we identified that any policy will help in reducing the "expected number of deaths", the "expected annual damage" in both Provinces, as well as the ecacuation cost and Rfr cost. Because every policy will decrease these variables, it is our task to find the most favourable and acceptable policy for the involved actors, regarding their interests and preferences. And that is where Many Objective Robust Optimization (MORO) comes in. MORO is a form of robust decision making that takes robustness into account in the search phase directly (Bartholomew & Kwakkel, 2020), by optimizing over robustness metrics rather than over the KPI's themselves.

As can be seen in the report, the aim of this analysis is to find a policy measure for Overijssel province that find is both robust and accepted by the different political parties. By using MORO, the robustness aspect can at least be achieved. And by taking into account the variance over locations, and by taking into account the most vulnerable locations, the acceptability part is aimed to be achieved.

MORO optimizes over the robustness-values generated by the self-defined KPIs of a policy. These robustness-values are said to be minimized or maximized in order to find an optimal policy within the space we limit by constraints and our definition of robustness.

### MORO Set up ###

As explained earlier, we use the signal-to-noise function (mean * std) to calculate the robustness values of a policy. This function will be used to minimize:

- The total expected number of deaths
- The total expected annual damage
- The total investment costs for dikes (later combined to 'Total non-RfR policy costs')
- The total evacuation costs (later combined to 'Total non RfR policy costs')
- The variance of costs between locations


Since the RfR costs are the same per scenario of a policy, there is no point in calculating the signal-to-noise of this variable. To solve that the optimization will not keep increasing the RfR costs, the aforementioned constraint is implemented. 

To calculate the total aggregated variables, all disaggregated variables are used as input. Only in this way, we can calculate the signal-to-noise of the aggregated variable.

```python
MINIMIZE = ScalarOutcome.MINIMIZE

percentile90 = functools.partial(np.percentile, q=90)

#Specifying robustness functions. All have to be minimized. All except for the variance will use the 'signal_to_noise' function.
#Variance will use the 'signal_to_noise_variation' function, which contains the extra step of first calculating the standard deviation.
robustnes_functions = [ScalarOutcome('90th percentile Death Number', kind=MINIMIZE, variable_name='Expected Number of Deaths',
                                      function=percentile90 ),
                        ScalarOutcome('90th percentile Annual Costs', kind=MINIMIZE, variable_name='Expected Annual Damage',
                                      function=percentile90 ),
                        ScalarOutcome('90th percentile Dike Costs', kind=MINIMIZE, variable_name='Dike Investment Costs',
                                      function=percentile90 ),
                        ScalarOutcome('90th percentile RfR Cost', kind=MINIMIZE, variable_name='RfR Investment Costs', 
                                      function=percentile90 ),
                        ScalarOutcome('90th percentile Evacuation', kind=MINIMIZE, variable_name='Evacuation Costs', 
                                      function=percentile90 )]
```

We made the uncertainty space smaller by implementing the ranges in which the worst-case scenarios occurred. Therefore, we made a decision to use a number of scenarios of 20. Since all of the scenarios are in the worst-case space, running 100 scenarios will only delay the run.

For the epsilon values, high values have been chosen. Since the signal-to-noise robustness function produces very large numbers and costs is a large number of itself, the epsilon values do not have to be very small. The exact values have been iteratively determined, by on one hand performing small runs and seeing the amount of solutions that came out of it. On the other hand, the table above that shows the robustness values for the 'random policy' run gives a rough insight in the produced values. Based on these values, epsilon values have been chosen that are not too small to not make any difference, but not too large to not be able to trade off between policies that have significantly different results. By making the epsilon values quite large, the run is also sped up.

```python
#Performing MORO: specifying the scenarios, the convergence metrics, the number of function evaluations and performing the robust_optimize function.
n_scenarios = 20
scenarios = sample_uncertainties(dike_model, n_scenarios)

convergence = [HyperVolume(minimum=[0,0,0,0,0], 
                           maximum=[death_25, annual_dmg_25, dike_25,rfr_25,evac_25 ]),
               EpsilonProgress()]

nfe = int(5000)

# with MultiprocessingEvaluator(dike_model) as evaluator:
#     archive, convergence = evaluator.robust_optimize(robustnes_functions, scenarios, 
#                                                nfe=nfe, convergence=convergence,
#                                                epsilons=[0.01]*len[robustnes_functions]
#                                                     )
# with open('results/MORORun.pkl', 'wb') as file_pi:
#     pickle.dump((archive, convergence), file_pi)
```

**Convergence**

In order to track whether the MORO has converged (i.e., no better solutions can be found within the demarcated policy space) to the possible optimum solutions, we analyse the convergence metrics. This will be done by analysing the epsilon-convergence and the hypervolume-convergence.

By looking at both the epsilon-progress and the hypervolume, we can conclude that both of them are convergenced. However, for the hypervolume, we may have picked ranges that were a bit too small (since the max value of 1 is reached). However, we assume that the epsilon-progress is the most important one and that the small ranges of hypervolume are not that big of a problem. Therefore, we can conclude the algorithm converged.  

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, sharex=True, figsize=(8, 4))
ax1.plot(convergence.nfe, convergence.epsilon_progress)
ax1.set_ylabel('$\epsilon$-progress')

ax1.set_xlabel('number of function evaluations')
ax2.set_xlabel('number of function evaluations')
plt.show()
```
